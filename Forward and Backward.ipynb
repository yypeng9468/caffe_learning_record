The backward pass begins with the loss and computes the gradient with respect to the output fWh. 



The gradient with respect to the rest of the model is computed layer-by-layer through the chain rule.



Layers with parameters, like the INNER_PRODUCT layer, compute the gradient with respect to their parameters fWWip during the backward step.

INNER_PRODUCT

The Net::Forward() and Net::Backward() methods carry out the respective passes while Layer::Forward() and Layer::Backward() compute each step.
