## backward pass

```python
img = mpimg.imread('./pics/backward.jpg')
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.axis('off')
plt.show()
```

The backward pass begins with the loss and computes the gradient with respect to the output ∂fW∂h. 

反向传播从损失出发，首先计算输出相对于损失的梯度。

The gradient with respect to the rest of the model is computed layer-by-layer through the chain rule.

模型的其他部分相对于损失的梯度是通过链式法则逐层计算的

Layers with parameters, like the INNER_PRODUCT layer, compute the gradient with respect to their parameters ∂fW∂Wip during the backward step.

那些有参数的层，比如INNER_PRODUCT层，在反向传播步骤中，会计算参数相对于损失的梯度。

The Net::Forward() and Net::Backward() methods carry out the respective passes while Layer::Forward() and Layer::Backward() compute each step.
